```python
import requests
from bs4 import BeautifulSoup
import math
import pandas as pd
import time
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

def get_with_retries(url, headers):
    session = requests.Session()
    retry = Retry(
        total=5,
        read=5,
        connect=5,
        backoff_factor=0.3,
        status_forcelist=(500, 502, 504),
    )
    adapter = HTTPAdapter(max_retries=retry)
    session.mount("http://", adapter)
    session.mount("https://", adapter)
    return session.get(url, headers=headers)

hot_keywords = ["Python", "Machine Learning", "Data Analysis", "SQL", "Statistics", "Big Data", "Deep Learning", "NLP", "AI", "Data Mining"]

l = []
o = {}
k = []
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
}
target_url = 'https://www.linkedin.com/jobs/search/?currentJobId=3942778965&geoId=101165590&keywords=data%20scientist&location=United%20Kingdom&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true'

for i in range(0, math.ceil(25/25)):
    res = get_with_retries(target_url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    alljobs_on_this_page = soup.find_all("li")
    print(len(alljobs_on_this_page))

    for x in range(0, len(alljobs_on_this_page)):
        base_card = alljobs_on_this_page[x].find("div", {"class": "base-card"})
        if base_card:
            data_entity_urn = base_card.get('data-entity-urn')
            if data_entity_urn:
                jobid = data_entity_urn.split(":")[3]
                l.append(jobid)
    
    time.sleep(1)  # Add delay to avoid rate limiting

target_url = 'https://www.linkedin.com/jobs-guest/jobs/api/jobPosting/{}'

for j in range(0, len(l)):
    try:
        resp = get_with_retries(target_url.format(l[j]), headers=headers)
        soup = BeautifulSoup(resp.text, 'html.parser')

        try:
            o["company"] = soup.find("div", {"class": "top-card-layout__card"}).find("a").find("img").get('alt')
        except AttributeError:
            o["company"] = None

        try:
            o["job-title"] = soup.find("div", {"class": "top-card-layout__entity-info"}).find("a").text.strip()
        except AttributeError:
            o["job-title"] = None

        try:
            o["level"] = soup.find("ul", {"class": "description__job-criteria-list"}).find("li").text.replace("Seniority level", "").strip()
        except AttributeError:
            o["level"] = None

        try:
            job_description = soup.find("div", {"class": "jobs-box__html-content"}).text.strip()
            o["job-description"] = job_description
        except AttributeError:
            o["job-description"] = None
            job_description = ""

        found_keywords = [keyword for keyword in hot_keywords if keyword.lower() in job_description.lower()]
        o["hot-keywords"] = ", ".join(found_keywords) if found_keywords else None

        k.append(o)
        o = {}

    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
        continue

    time.sleep(1)  # Add delay to avoid rate limiting

df = pd.DataFrame(k)
df.to_csv('linkedinjobs3.csv', index=False, encoding='utf-8')
print(k)

```
